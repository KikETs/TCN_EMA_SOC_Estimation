# Generate manuscript and SI tables

Each output cell runs the matching script and displays the generated artifact directly in the notebook.

```python
from pathlib import Path
import subprocess
import pandas as pd
from IPython.display import display

ROOT = Path.cwd()
if ROOT.name in {'Figures', 'Tables'}:
    ROOT = ROOT.parent

pd.set_option('display.max_columns', 80)
pd.set_option('display.width', 160)

def run_script(script_name):
    script = ROOT / 'Scripts' / script_name
    subprocess.run(['python', str(script)], cwd=ROOT, check=True)

def run_table(label, script_name, csv_name):
    run_script(script_name)
    path = ROOT / 'Tables' / csv_name
    df = pd.read_csv(path)
    print(label)
    display(df)
```
