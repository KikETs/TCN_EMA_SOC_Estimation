# Generate manuscript and SI figures

Each output cell runs the matching script and displays the generated artifact directly in the notebook.

```python
from pathlib import Path
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ROOT = Path.cwd()
if ROOT.name in {'Figures', 'Tables'}:
    ROOT = ROOT.parent

plt.rcParams['figure.dpi'] = 120

def run_script(script_name):
    script = ROOT / 'Scripts' / script_name
    subprocess.run(['python', str(script)], cwd=ROOT, check=True)

def run_figure(label, script_name, png_name):
    run_script(script_name)
    path = ROOT / 'Figures' / png_name
    img = mpimg.imread(path)
    height, width = img.shape[:2]
    fig_w = min(13.0, max(6.0, width / 160.0))
    fig_h = min(14.0, max(3.0, height / 160.0))
    fig, ax = plt.subplots(figsize=(fig_w, fig_h))
    ax.imshow(img)
    ax.axis('off')
    plt.show()
```
